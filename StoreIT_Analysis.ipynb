```python
import folium
from IPython.display import display

# Base map centered on Singapore
m = folium.Map(location=[1.3521, 103.8198], zoom_start=11)

# Loop through each row and add a marker with tooltip and popup
for _, row in hosts.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        tooltip=row['Username'],  
        icon=folium.Icon(color='blue', icon='home')
    ).add_to(m)

display(m)
```

```python
import pandas as pd
import plotly.express as px

# Assuming 'hosts' DataFrame is already loaded

# 1. Convert dates to datetime
hosts['Available Start Date'] = pd.to_datetime(hosts['Available Start Date'])
hosts['Available End Date'] = pd.to_datetime(hosts['Available End Date'])

# 2. Optional: Cap extreme future dates (e.g., 2045) to improve visualization
max_allowed_date = pd.to_datetime("2027-12-31")
hosts['Available End Date'] = hosts['Available End Date'].apply(
    lambda x: min(x, max_allowed_date)
)

# 3. Optional: Sort hosts by start date (for neater timeline)
hosts = hosts.sort_values(by='Available Start Date')

# 4. Plot Gantt chart
fig = px.timeline(
    hosts,
    x_start="Available Start Date",
    x_end="Available End Date",
    y="Username",
    title="Host Availability Timeline",
    labels={"Username": "Host"},
)

fig.update_yaxes(autorange="reversed")  # Earliest host on top
fig.update_layout(
    height=600,
    xaxis_title="Date",
    yaxis_title="Host",
    margin=dict(l=150, r=20, t=50, b=20)
)

fig.show()
```
